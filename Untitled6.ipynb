```python
logs = [
    "2026-01-08 | INFO | alice | Login",
    "2026-01-08 | ERROR | bob | Wrong password",
    "2026-01-08 | WARNING | alice | Multiple attempts",
    "2026-01-08 | ERROR | alice | Account locked",
    "2026-01-08 | INFO | bob | Logout"
]

parsed_logs = []

# -------- PARSE LOG ENTRIES --------
for log in logs:
    parts = log.split(" | ")
    parsed_logs.append(parts)

# -------- MENU --------
while True:
    print("\n--- Log File Analyzer ---")
    print("1. Count events by severity")
    print("2. Count error types")
    print("3. Analyze user activity")
    print("4. Filter logs by severity")
    print("5. Exit")

    choice = input("Enter choice: ")

    # ---- OPTION 1 ----
    if choice == "1":
        severity_count = {}
        for p in parsed_logs:
            severity = p[1]
            severity_count[severity] = severity_count.get(severity, 0) + 1
        print(severity_count)

    # ---- OPTION 2 ----
    elif choice == "2":
        error_count = {}
        for p in parsed_logs:
            if p[1] == "ERROR":
                error_count[p[3]] = error_count.get(p[3], 0) + 1
        print(error_count)

    # ---- OPTION 3 ----
    elif choice == "3":
        user_count = {}
        for p in parsed_logs:
            user = p[2]
            user_count[user] = user_count.get(user, 0) + 1
        print(user_count)

    # ---- OPTION 4 ----
    elif choice == "4":
        level = input("Enter severity (INFO / WARNING / ERROR): ")
        for p in parsed_logs:
            if p[1] == level:
                print(p)

    # ---- EXIT ----
    elif choice == "5":
        print("Exiting program")
        break

    else:
        print("Invalid choice")
```

```python
# ----- ROOMS AND CAPACITY -----
rooms = ["R1", "R2", "R3"]
capacity = [2, 2, 1]

# ----- STUDENT ALLOCATIONS -----
students = []
room_allotted = []

# ----- MENU -----
while True:
    print("\n--- Hostel Room Allocation System ---")
    print("1. Allocate Room")
    print("2. Vacate Room")
    print("3. View Occupancy")
    print("4. Allocation Report")
    print("5. Exit")

    choice = input("Enter choice: ")

    # ----- ALLOCATE ROOM -----
    if choice == "1":
        name = input("Enter student name: ")

        if name in students:
            print("Student already has a room.")
        else:
            allocated = False
            for i in range(len(rooms)):
                count = room_allotted.count(rooms[i])
                if count < capacity[i]:
                    students.append(name)
                    room_allotted.append(rooms[i])
                    print("Room", rooms[i], "allocated to", name)
                    allocated = True
                    break

            if not allocated:
                print("No rooms available.")

    # ----- VACATE ROOM -----
    elif choice == "2":
        name = input("Enter student name: ")

        if name in students:
            index = students.index(name)
            print("Room", room_allotted[index], "vacated for", name)
            students.pop(index)
            room_allotted.pop(index)
        else:
            print("Student not found.")

    # ----- VIEW OCCUPANCY -----
    elif choice == "3":
        for r in rooms:
            print(r, ":", room_allotted.count(r), "students")

    # ----- ALLOCATION REPORT -----
    elif choice == "4":
        for i in range(len(students)):
            print(students[i], "->", room_allotted[i])

    # ----- EXIT -----
    elif choice == "5":
        print("Exiting system.")
        break

    else:
        print("Invalid choice.")
```
